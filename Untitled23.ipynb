```python
import numpy as np
from scipy.fftpack import fftn, fftshift
from scipy.integrate import simps

def shell_spectrum_1d(v, dx):
    # Compute Fourier transform of vector field
    vf = fftshift(fftn(v))
    
    # Compute wavevector magnitudes
    n = v.shape[0]
    k = np.fft.fftfreq(n, dx) * 2 * np.pi
    kx, ky, kz = np.meshgrid(k, k, k, indexing='ij')
    kmag = np.sqrt(kx**2 + ky**2 + kz**2)
    
    # Compute power spectrum and integrate over spherical shells
    bins = np.arange(kmag.max())
    indices = np.digitize(kmag.flat, bins)
    pspec = np.zeros(len(bins))
    count = np.zeros(len(bins))
    for i in range(1, len(bins)):
        shell = (indices == i).reshape(vf.shape[0], vf.shape[1], vf.shape[2])
        if shell.sum() == 0:
            continue
        pspec[i-1] = simps(np.abs(vf[shell])**2, dx=dx)
        count[i-1] = shell.sum()
    pspec /= count
    
    return bins, pspec
```

```python
def get_1d_coords(cmin=0, cmax=1, ndim=10, cell_centred=True):
    if cell_centred:
        d = (cmax-cmin) / np.float(ndim)
        offset = d/2
    else:
        d = (cmax-cmin) / np.float(ndim-1)
        offset = 0.0
    return np.linspace(cmin+offset, cmax-offset, num=ndim)

def get_2d_coords(cmin=[0,0], cmax=[1,1], ndim=[10,10], cell_centred=True):
    cmin = np.array(cmin)
    cmax = np.array(cmax)
    ndim = np.array(ndim)
    if cmin.ndim != 1: cmin = [cmin,cmin]
    if cmax.ndim != 1: cmax = [cmax,cmax]
    if ndim.ndim != 1: ndim = [ndim,ndim]
    c0 = get_1d_coords(cmin=cmin[0], cmax=cmax[0], ndim=ndim[0], cell_centred=cell_centred)
    c1 = get_1d_coords(cmin=cmin[1], cmax=cmax[1], ndim=ndim[1], cell_centred=cell_centred)
    return np.array(np.meshgrid(c0, c1, indexing='ij'))

def get_3d_coords(cmin=[0,0,0], cmax=[1,1,1], ndim=[10,10,10], cell_centred=True):
    cmin = np.array(cmin)
    cmax = np.array(cmax)
    ndim = np.array(ndim)
    if cmin.ndim != 1: cmin = [cmin,cmin,cmin]
    if cmax.ndim != 1: cmax = [cmax,cmax,cmax]
    if ndim.ndim != 1: ndim = [ndim,ndim,ndim]
    c0 = get_1d_coords(cmin=cmin[0], cmax=cmax[0], ndim=ndim[0], cell_centred=cell_centred)
    c1 = get_1d_coords(cmin=cmin[1], cmax=cmax[1], ndim=ndim[1], cell_centred=cell_centred)
    c2 = get_1d_coords(cmin=cmin[2], cmax=cmax[2], ndim=ndim[2], cell_centred=cell_centred)
    return np.array(np.meshgrid(c0, c1, c2, indexing='ij'))
# === bin data with bin_values (same size as data) into bins (number or array of bin edges) ===
def get_binned_stats(data, bin_values, bins=None, statistic='mean', **kwargs):
    from scipy.stats import binned_statistic
    binned_stats = binned_statistic(bin_values.flatten(), data.flatten(), bins=bins, statistic=statistic)
    return binned_stats.statistic, binned_stats.bin_edges

def get_coords(cmin, cmax, ndim, cell_centred=True):
    if (type(cmin) != list) and (type(cmin) != np.ndarray): print("need list or nump array inputs", error=True)
    cmin = np.array(cmin)
    cmax = np.array(cmax)
    ndim = np.array(ndim)
    if (cmin.shape != cmax.shape) or (cmax.shape != cmax.shape):
        print("Error: cmin, cmax, ndim, all must have the same shape.", error=True)
    if ndim.shape[0] == 1: return np.array(get_1d_coords(cmin[0], cmax[0], ndim[0], cell_centred))
    if ndim.shape[0] == 2: return np.array(get_2d_coords(cmin, cmax, ndim, cell_centred))
    if ndim.shape[0] == 3: return np.array(get_3d_coords(cmin, cmax, ndim, cell_centred))


def get_spectrum(data_in, ncmp=1):
    data = np.copy(data_in) # flatten to strip any possible extra dimensions
    if (ncmp == 1) and (data.shape[0] > 1):
        data = np.array([data]) # add an extra (fake) index, so we can index as if there were components
    num = np.array(data[0].shape) # number of points in data
    ndim = len(num) # number of dimensions
    ks = -num//2 # k start
    ke = np.array([np.min([-ks[d], num[d]//2-1]) for d in range(ndim)]) # k end
    k = get_coords(ks, ke, num, cell_centred=False) # get k vector with k=0 in the center
    if ndim == 1: k_abs = np.abs(k) # length of k vector
    if ndim  > 1: k_abs = np.sqrt((k**2).sum(axis=0)) # length of k vector
    bins = np.arange(np.max(num)//2) - 0.5 # k bins for 1D spectrum
    data_ft = []
    for d in range(ncmp):
        data_ft.append(np.fft.fftn(data[d], norm='forward')) # FFT
        data_ft[d] = np.fft.fftshift(data_ft[d]) # shift k=0 to center
    data_ft = np.array(data_ft)
    # get total power
    power_tot = (np.abs(data_ft)**2).sum(axis=0)
    # Helmholtz decomposition
    if ncmp > 1: # there is more than 1 component
        power_lgt = np.zeros(num, dtype=complex)
        if ndim == 1: power_lgt += k*data_ft[0] # 1D case: scalar product (k is a 1D array and we only use x-component data for the longitudinal power)
        if ndim >= 2: # 2D and 3D cases: scalar product (get power along k); if ndim < ncmp (i.e., 2.5D), the z-component does not enter the scalar product
            for d in range(ndim): power_lgt += k[d]*data_ft[d].T # scalar product
        power_lgt = np.abs(power_lgt/np.maximum(k_abs,1e-99))**2
        power_trv = power_tot - power_lgt
        print("tot power = "+str(power_tot.sum()/ncmp))
        print("lgt power = "+str(power_lgt.sum()/ncmp)+", relative to tot: "+str(power_lgt.sum()/power_tot.sum()))
        print("trv power = "+str(power_trv.sum()/ncmp)+", relative to tot: "+str(power_trv.sum()/power_tot.sum()))
    # bin in k shells
    spect_tot, bins = get_binned_stats(power_tot, k_abs, bins=bins)
    bin_centers = bins[:-1]+0.5
    integral_factor = bin_centers**(ndim-1)
    if ndim > 1: integral_factor *= np.pi*2*(ndim-1)
    ret = {'k': bin_centers, 'P_tot': spect_tot*integral_factor}
    if ncmp > 1:
        spect_lgt, bins = get_binned_stats(power_lgt, k_abs, bins=bins)
        spect_trv, bins = get_binned_stats(power_trv, k_abs, bins=bins)
        ret['P_lgt'] = spect_lgt*integral_factor
        ret['P_trv'] = spect_trv*integral_factor
    # return dict
    return ret
```

```python
import numpy as np
from numpy.fft import fftn, fftshift
from itertools import product

def shell_spectrum_1d(field, boxsize):
    # Compute Fourier transform of field.
    ft_field = np.fft.fftn(field) / field.size
    ft_field = np.fft.fftshift(ft_field)

    # Compute the k-space coordinates of the Fourier modes.
    k = np.fft.fftfreq(ft_field.shape[0], d=boxsize/ft_field.shape[0])
    k_coords = np.array(np.meshgrid(k, k, k, indexing='ij'))

    # Compute the magnitude of the wavevector.
    k_magnitude = np.sqrt(np.sum(k_coords**2, axis=0))

    # Bin the wavevectors into radial shells.
    num_bins = int(np.sqrt(3) * ft_field.shape[0] / 2)
    k_bins = np.linspace(0, np.sqrt(3)*ft_field.shape[0]/2, num_bins+1)
    digitized = np.digitize(k_magnitude, k_bins)
    
    # Compute the spherically-averaged power spectrum.
    power_spectrum = np.abs(ft_field)**2
    bin_means = [np.mean(power_spectrum[digitized == i]) for i in range(1, len(k_bins))]
    bin_counts = [np.sum(digitized == i) for i in range(1, len(k_bins))]
    bin_means = np.array(bin_means)
    bin_counts = np.array(bin_counts)
    bin_means = bin_means / bin_counts
    
    # Compute the central wavenumber of each radial shell.
    k_perp = 0.5 * (k_bins[1:] + k_bins[:-1])

    return k_perp, bin_means

```

```python
import numpy as np
from numpy.fft import fftn, ifftn

def shell_to_shell_transfer_function(u, b):
    """
    Computes the shell-to-shell spectral energy transfer function between
    velocity and magnetic fields.

    Parameters:
    u (ndarray): 3D array of velocity field.
    b (ndarray): 3D array of magnetic field.

    Returns:
    T (ndarray): 2D array of shell-to-shell spectral energy transfer function.
    """

    # Compute the Fourier transforms of the velocity and magnetic fields
    uf = fftn(u)
    bf = fftn(b)

    # Compute the wavevectors
    nx, ny, nz = u.shape
    kx = np.fft.fftfreq(nx, d=1)
    ky = np.fft.fftfreq(ny, d=1)
    kz = np.fft.fftfreq(nz, d=1)
    kx, ky, kz = np.meshgrid(kx, ky, kz, indexing='ij')
    k = np.sqrt(kx**2 + ky**2 + kz**2)

    # Compute the shell-to-shell spectral energy transfer function
    T = np.zeros((nx, ny, nz))
    for i in range(nx):
        for j in range(ny):
            for l in range(nz):
                for i2 in range(nx):
                    for j2 in range(ny):
                        for l2 in range(nz):
                            k1 = k[i, j, l]
                            k2 = k[i2, j2, l2]
                            if k1 > 0 and k2 > 0:
                                T[i, j, l] += -1j * (uf[i, j, l].conjugate() * bf[i2, j2, l2] / k2 +
                                                    bf[i, j, l].conjugate() * uf[i2, j2, l2] / k2)

    return T
```

```python
import numpy as np
from numpy.fft import fftn, ifftn

def shell_to_shell_mag_energy_transfer_function(u, b):
    """
    Computes the shell-to-shell energy transfer function for the magnetic energy
    from the velocity.

    Parameters:
    u (ndarray): 3D array of velocity field.
    b (ndarray): 3D array of magnetic field.

    Returns:
    T (ndarray): 2D array of shell-to-shell energy transfer function for the magnetic energy.
    """

    # Compute the Fourier transforms of the velocity and magnetic fields
    uf = fftn(u)
    bf = fftn(b)

    # Compute the wavevectors
    nx, ny, nz = u.shape
    kx = np.fft.fftfreq(nx, d=1)
    ky = np.fft.fftfreq(ny, d=1)
    kz = np.fft.fftfreq(nz, d=1)
    kx, ky, kz = np.meshgrid(kx, ky, kz, indexing='ij')
    k = np.sqrt(kx**2 + ky**2 + kz**2)

    # Compute the shell-to-shell energy transfer function for the magnetic energy
    T = np.zeros((nx, ny, nz))
    for i in range(nx):
        for j in range(ny):
            for l in range(nz):
                for i2 in range(nx):
                    for j2 in range(ny):
                        for l2 in range(nz):
                            k1 = k[i, j, l]
                            k2 = k[i2, j2, l2]
                            if k1 > 0 and k2 > 0:
                                T[i, j, l] += -1j * bf[i, j, l].conjugate() * uf[i2, j2, l2]

    return T
```

```python
import numpy as np

def compute_transfer_functions(vel_spectrum, mag_spectrum, k, k_max):
    """
    Compute shell-to-shell energy transfer functions between magnetic and velocity fields.

    :param vel_spectrum: numpy array with the Fourier coefficients of the velocity field.
    :param mag_spectrum: numpy array with the Fourier coefficients of the magnetic field.
    :param k: numpy array with the wavenumbers corresponding to the Fourier coefficients.
    :param k_max: maximum wavenumber to consider for the transfer function calculation.

    :return: numpy array with the transfer functions for each shell.
    """
    # Calculate the energy spectra for the velocity and magnetic fields
    vel_energy_spectrum = 0.5 * (np.abs(vel_spectrum)**2)
    mag_energy_spectrum = 0.5 * (np.abs(mag_spectrum)**2)

    # Compute the shell-averaged spectra for the velocity and magnetic fields
    vel_shell_spectrum = np.zeros(k_max)
    mag_shell_spectrum = np.zeros(k_max)
    for i in range(k_max):
        shell_mask = (k >= i) & (k < i+1)
        vel_shell_spectrum[i] = np.mean(vel_energy_spectrum[shell_mask])
        mag_shell_spectrum[i] = np.mean(mag_energy_spectrum[shell_mask])

    # Compute the transfer functions for each shell
    transfer_functions = np.zeros(k_max)
    for i in range(1, k_max):
        vel_energy_in = np.sum(vel_energy_spectrum[k >= i-1])
        vel_energy_out = np.sum(vel_energy_spectrum[k >= i])
        mag_energy_in = np.sum(mag_energy_spectrum[k >= i-1])
        mag_energy_out = np.sum(mag_energy_spectrum[k >= i])
        transfer_functions[i] = (
            (vel_energy_out - vel_energy_in) * mag_shell_spectrum[i]
            - vel_shell_spectrum[i] * (mag_energy_out - mag_energy_in)
        )

    return transfer_functions
```

```python
import numpy as np

# Generate random velocity and magnetic fields in Fourier space
n = 256
k = np.arange(n)
vel_spectrum = np.random.randn(n) + 1j*np.random.randn(n)
mag_spectrum = np.random.randn(n) + 1j*np.random.randn(n)
```

```python
# Calculate the transfer functions
k_max = n//2
transfer_functions = compute_transfer_functions(vel_spectrum, mag_spectrum, k, k_max)
```

```python
import matplotlib.pyplot as plt

# Plot the transfer functions
plt.plot(np.arange(1, k_max), transfer_functions[1:])
plt.xlabel('Spherical shell index')
plt.ylabel('Shell-to-shell transfer function')
plt.title('Transfer functions between velocity and magnetic fields')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random velocity and magnetic fields
n = 256
k = np.arange(1, n+1)
vel_spectrum = np.random.randn(n) + 1j*np.random.randn(n)
mag_spectrum = np.random.randn(n) + 1j*np.random.randn(n)

# Compute transfer functions
k_max = n//2
transfer_functions = compute_transfer_functions(vel_spectrum, mag_spectrum, k, k_max)

# Plot transfer functions
plt.figure(figsize=(8, 6))
plt.plot(k[1:k_max], transfer_functions[1:k_max+1])
plt.xscale('log')
plt.yscale('symlog', linthresh=1e-6)
plt.xlabel('Wavenumber (k)')
plt.ylabel('T(k)')
plt.title('Shell-to-shell energy transfer between magnetic and velocity fields')
plt.show()
```
